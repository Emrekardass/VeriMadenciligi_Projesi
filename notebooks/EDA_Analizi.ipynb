```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Ürettiğimiz veriyi klasörden okuyalım
df = pd.read_csv("../data/sensor_data.csv")

# 2. Veri setinin ilk 5 satırına ve genel yapısına bakalım
display(df.head())
display(df.info())

# 3. Hızlıca bir korelasyon matrisi çizdirelim
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Jupyter Notebook Üzerinden EDA Kontrolü")
plt.show()
```
